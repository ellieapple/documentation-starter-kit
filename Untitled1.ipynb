## Input Markdown

### Subtask:
Create a method for the user to input markdown text into the application.


**Reasoning**:
First, I will import the `ipywidgets` library to create interactive input elements. This is a prerequisite for creating the markdown input widget.



```python
import ipywidgets as widgets
from IPython.display import display

print("ipywidgets imported successfully.")
```

**Reasoning**:
Now that `ipywidgets` is imported, I will create the `Textarea` widget for markdown input, configure its placeholder, description, and layout as instructed, and then display it to the user.



```python
markdown_input = widgets.Textarea(
    placeholder='Enter your Markdown text here...',
    description='Markdown Input:',
    layout=widgets.Layout(width='800px', height='300px')
)
display(markdown_input)
```

## Convert Markdown to Text

### Subtask:
Write Python code to convert the input markdown text into plain text. This will involve using the `markdown` and `beautifulsoup` libraries to strip markdown formatting.


**Reasoning**:
First, I will install the necessary libraries, `markdown` and `beautifulsoup4`, using pip. This is a prerequisite for converting markdown to plain text.



```python
pip install markdown beautifulsoup4
```

**Reasoning**:
Now that the necessary libraries are installed, I will import them and define the `markdown_to_plain_text` function as instructed. This function will encapsulate the logic for converting markdown to HTML and then stripping the HTML tags to get plain text.



```python
import markdown
from bs4 import BeautifulSoup

def markdown_to_plain_text(markdown_string):
    """Converts a markdown string to plain text by first converting to HTML and then stripping tags."""
    # Convert markdown to HTML
    html = markdown.markdown(markdown_string)

    # Parse HTML and extract plain text
    soup = BeautifulSoup(html, 'html.parser')
    plain_text = soup.get_text()

    return plain_text

print("The 'markdown_to_plain_text' function has been defined.")
```

## Convert Markdown to Text

### Subtask:
Convert the input markdown text into plain text using the defined function and display the result.


**Reasoning**:
I will implement the conversion logic by creating an output widget, a button, and a function that handles the conversion from the `markdown_input` widget to plain text using the previously defined `markdown_to_plain_text` function. Finally, I will display the button and the output widget.



```python
output_widget = widgets.Output()
convert_button = widgets.Button(description='Convert to Plain Text')

def on_convert_button_clicked(b):
    with output_widget:
        output_widget.clear_output()
        markdown_text = markdown_input.value
        plain_text = markdown_to_plain_text(markdown_text)
        print(plain_text)

convert_button.on_click(on_convert_button_clicked)

display(convert_button, output_widget)
print("Conversion button and output widget displayed. Enter markdown in the input box and click 'Convert to Plain Text'.")
```

## Provide Download Option

### Subtask:
Generate a download link or a mechanism for the user to download the converted plain text as a `.txt` file.


**Reasoning**:
First, I will import the `HTML` class from `IPython.display` as instructed, which is necessary for displaying the download link.



```python
from IPython.display import HTML
import base64

print("HTML and base64 imported successfully.")
```

**Reasoning**:
Now that the necessary modules are imported, I will modify the `on_convert_button_clicked` function to generate a download link after converting the markdown to plain text. This involves encoding the plain text in base64 and creating an HTML `<a>` tag with a data URL for download.



```python
output_widget = widgets.Output()
convert_button = widgets.Button(description='Convert to Plain Text')

def on_convert_button_clicked(b):
    with output_widget:
        output_widget.clear_output()
        markdown_text = markdown_input.value
        plain_text = markdown_to_plain_text(markdown_text)

        # Display the plain text
        print(plain_text)
        print("\n") # Add a newline for better separation

        # Generate download link
        b64_text = base64.b64encode(plain_text.encode()).decode()
        download_filename = "converted_text.txt"
        href = f"<a href='data:text/plain;base64,{b64_text}' download='{download_filename}'>Download Plain Text</a>"
        display(HTML(href))

convert_button.on_click(on_convert_button_clicked)

display(convert_button, output_widget)
print("Conversion button, output widget, and download link functionality displayed. Enter markdown in the input box and click 'Convert to Plain Text'.")
```
